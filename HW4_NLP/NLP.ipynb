```python
import spacy
nlp = spacy.load('en_core_web_sm') # the data corpus to load based on which its model's intelligence will be dependent on
doc = nlp("Apple is looking at buying U.K. startup for $1 billion.")
for token in doc:
  print(token.text)
```

```python
notes = []
with open('notes_headache.txt', 'r') as fin:
  lines = fin.readlines()
  for line in lines:
    notes.append(line)
print(notes)
print(len(notes))
```

```python
doc = []
for i in range(len(notes)):
  doc.append(nlp(notes[i]))
  print(doc[-1])
  print('*************************************************************************************************************')
```

```python
# Token
token_without_punct = []
for i in range(len(doc)):
  token_without_punct.append([token.orth_ for token in doc[i] if not token.is_punct | token.is_space])
  print(token_without_punct[-1])
  print('*******************************************************************************************')
```

```python
# Lemmatization
# It helps with plural -> singular and all verbe tenses -> verb
doc = nlp("Apples and oranges are similar. Boots and hippos aren't.")
for token in doc:
  print(token, token.lemma, token.lemma_)
```

```python
# POS Tagging
doc = nlp("Apple is looking for buying U.K. startup for $1 billion.")
for token in doc:
  print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_, token.shape_, token.is_alpha, token.is_stop)
```

```python
# [WARNING] In Lecture, Prof Ding re-runs the code block above where `headache_notes.txt`
#           is loaded into the variable `doc`——before running this cell. Count 4 cells up.

for i in range(len(doc)):
  for token in doc[i]:
    print(token.text, token.pos_)
  print('**************************************************')
```

```python
# Name Entities
# they idenitfy real-world objects like a person, country , product or book title.
doc = nlp("Apple is looking at buying U.K. startup for #1 billion.")
for ent in doc.ents:
  print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
# for the medical note, the model doesnt do that well since it doesnt pick things like
# headache, back of head tec
doc = []
for i in range(len(notes)):
  doc.append(nlp(notes[i]))
  for ent in doc[-1].ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
  print("*********************************************************")
```

```python
# Entity Visualizer
from spacy import displacy
for i in range(len(doc)):
  displacy.render(doc[i], style="ent", jupyter=True)
  print('*********************************************************************************************************************************************************************')
```

```python
#sentence identifier
for i in range(len(doc)):
  for ix, sent in enumerate(doc[i].sents, 1):
    print("Sentence number {}:{}".format(ix, sent))
  print('*******************************************************************************************')
```

```python
# dependence tree
# for i in range(len(doc)):
i = 0
sentence_spans = list(doc[i].sents)
displacy.render(sentence_spans, style="dep", jupyter=True)
```

```python
# Spacy has its own vocabulary where each word is mapped to a hash
doc = nlp("I love apple")
for word in doc:
  lexeme = doc.vocab[word.text]
  print(lexeme.text, lexeme.orth, lexeme.shape_, lexeme.prefix_, lexeme.suffix_, lexeme.is_alpha, lexeme.is_digit, lexeme.is_title, lexeme.lang_)
```
